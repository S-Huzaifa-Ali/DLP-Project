```python
import threading
import time
import os
import shutil

def auto_save_weights():
    while True:
        time.sleep(300)
        try:
            weights_path = '/kaggle/working/Yolo_Food_Detection/run1/weights'
            if os.path.exists(f'{weights_path}/last.pt'):
                shutil.copy(f'{weights_path}/last.pt', '/kaggle/working/last_backup.pt')
                print(f"Auto-saved at {time.strftime('%H:%M:%S')}")
        except:
            pass

t = threading.Thread(target=auto_save_weights, daemon=True)
t.start()
print("Auto-save started!")
```

```python
from ultralytics import YOLO

model = YOLO('yolov8m.pt')

model.train(
    data='/kaggle/working/food_detection.yaml',
    epochs=50,
    imgsz=640,
    batch=32,
    device='0,1',
    workers=4,
    project='/kaggle/working/Yolo_Food_Detection',
    name='run1',
    patience=10,
    save=True,
    plots=True,
    save_period=2,
)
```

```python
import shutil
import os

weights = '/kaggle/working/Yolo_Food_Detection/run1/weights'

shutil.copy(f'{weights}/best.pt', '/kaggle/working/yolo_best.pt')
shutil.copy(f'{weights}/last.pt', '/kaggle/working/yolo_last.pt')

print("Done!")
print(os.listdir('/kaggle/working'))
```
